```python
import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import MinMaxScaler
from sklearn import tree
from sklearn import linear_model


file = "train.csv"
data = pd.read_csv(file)
data['hh_totalmembers'].describe()
data = data.drop('surveyid', axis=1)
data = data.drop('survey_date', axis=1)


n_records = data.shape[0]
n_depressed = data['depressed'].value_counts()[1]

print("Total number of records: {}".format(n_records))
print("Total number of people depressed: {}".format(n_depressed))

```

```python
## Making categorical variables into one-hot vector
cat_variables = ['village','femaleres','married','labor_primary','day_of_week']

one_hot = pd.get_dummies(data, columns = cat_variables)
one_hot.head(n=5)

var_names = list(one_hot) #keeping track of variable names
```

```python
## Taking out variables that have more than 500 missing values
listy = one_hot.isnull().sum()
names = list(listy.index)
to_take_out = []
for element in names:
    if listy[element] >=500:
        to_take_out.append(element)
one_hot = one_hot.drop(to_take_out, axis=1)
```

```python
#Changing variables with less than 200 into their most frequent value

listy = one_hot.isnull().sum()
names = list(listy.index)
few_missing = []
for element in names:
    if listy[element] > 0 and listy[element] <200:
        few_missing.append(element)


imp = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
imp.fit(one_hot[few_missing])
one_hot[few_missing]= imp.transform(one_hot[few_missing])
one_hot = pd.DataFrame(one_hot)
```

```python
listy = one_hot.isnull().sum()
names = list(listy.index)
missing = []
for element in names:
    if listy[element] > 0:
        missing.append(element)
missing
```

```python
var_names = list(one_hot)
imp = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
imp.fit(one_hot)
temp_data = imp.transform(one_hot)
temp_data = pd.DataFrame(temp_data)
temp_data.columns = var_names

X = temp_data.drop('hh_totalmembers',axis = 1)
y = temp_data['hh_totalmembers']
lm = linear_model.LinearRegression()
new_y = lm.fit(X,y).predict(X)
```

```python
one_hot_copy = one_hot.copy()


for variable in missing:
    X = temp_data.drop(variable,axis = 1)
    y = temp_data[variable]
    lm = linear_model.LinearRegression()
    new_y = lm.fit(X,y).predict(X)
    for value in range(1143):
        if one_hot[variable][value]!= one_hot[variable][value]:
            one_hot_copy[variable][value] = new_y[value]
```

```python
one_hot_copy.isnull().sum()
```

```python
## Scaling values
scaler = MinMaxScaler()
var_names = list(one_hot_copy)
train = scaler.fit_transform(one_hot_copy)
train = pd.DataFrame(train)
train.columns = var_names
display(train.head(n = 5))

X_train = train.drop('depressed', axis=1)
Y_train = train['depressed']


data = pd.read_csv('test.csv')
data = data.drop('surveyid', axis=1)
data = data.drop('survey_date', axis=1)
data['age'] = data['age'].replace({'.d':np.nan})


data = data.drop(to_take_out, axis=1)
one_hot = pd.get_dummies(data, columns = cat_variables)
var_names = list(one_hot) #keeping track of variable names
## Imputing missing values
imp = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
imp.fit(one_hot)
one_hot= imp.transform(one_hot)
test = pd.DataFrame(one_hot)
## Scaling values
scaler = MinMaxScaler()
test = scaler.fit_transform(test)
test = pd.DataFrame(test)
test.columns = var_names
Y_test = test['depressed']
X_test = test.drop('depressed', axis=1)


train_list = list(X_train)
test_list = list(X_test)
listy = []
for element in train_list:
    if element not in test_list:
        listy.append(element)
        X_test[element] = 0

        
for element in test_list:
    if element not in train_list:
        listy.append(element)
        X_train[element] = 0


        
clf = tree.DecisionTreeClassifier(min_samples_leaf=16)


clf.fit(X_train, Y_train)
clf.score(X_test,Y_test)

Y_predicted = clf.predict(X_test)

tots = 0
for i in range(len(Y_test)):
    if Y_predicted[i] != Y_test[i]:
        tots +=1
print("The percentage of predictions that are not the same is {:.2f}%".format(tots/len(Y_predicted)*100))



ids = pd.read_csv('test.csv')
ids = ids['surveyid']
ids = pd.DataFrame(ids)
Y_predicted = pd.DataFrame(Y_predicted)
Y_predicted.columns = ['depressed']

sub = pd.merge(ids,Y_predicted,left_index =True,right_index=True)

sub['depressed'] = sub['depressed'].astype(int)
sub.to_csv('andres_parrado_submission6.csv', encoding='utf-8')
```
