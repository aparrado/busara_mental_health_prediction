```python
import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Ridge
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression


file = "train.csv"
data = pd.read_csv(file)
data['hh_totalmembers'].describe()
data = data.drop('surveyid', axis=1)
data = data.drop('survey_date', axis=1)
data.head(n=5)

```

```python
n_records = data.shape[0]
n_depressed = data['depressed'].value_counts()[1]

print("Total number of records: {}".format(n_records))
print("Total number of people depressed: {}".format(n_depressed))
```

```python
## Making categorical variables into one-hot vector
cat_variables = ['village','femaleres','married','labor_primary','day_of_week']

one_hot = pd.get_dummies(data, columns = cat_variables)
one_hot.head(n=5)

var_names = list(one_hot) #keeping track of variable names
```

```python
## Imputing missing values
imp = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
imp.fit(one_hot)
one_hot= imp.transform(one_hot)
train = pd.DataFrame(one_hot)
train.columns = var_names
train.head(n=5)
#train.isnull().sum()
```

```python
## Scaling values
scaler = MinMaxScaler()
train = scaler.fit_transform(train)
train = pd.DataFrame(train)
train.columns = var_names
display(train.head(n = 5))
```

```python
X_train = train.drop('depressed', axis=1)
Y_train = train['depressed']


data = pd.read_csv('test.csv')
data = data.drop('surveyid', axis=1)
data = data.drop('survey_date', axis=1)
data['age'] = data['age'].replace({'.d':np.nan})
one_hot = pd.get_dummies(data, columns = cat_variables)
var_names = list(one_hot) #keeping track of variable names
## Imputing missing values
imp = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
imp.fit(one_hot)
one_hot= imp.transform(one_hot)
test = pd.DataFrame(one_hot)
## Scaling values
scaler = MinMaxScaler()
test = scaler.fit_transform(test)
test = pd.DataFrame(test)
test.columns = var_names
Y_test = test['depressed']
X_test = test.drop('depressed', axis=1)


train_list = list(X_train)
test_list = list(X_test)
listy = []
for element in train_list:
    if element not in test_list:
        listy.append(element)
        X_test[element] = 0

        
for element in test_list:
    if element not in train_list:
        listy.append(element)
        X_train[element] = 0

```

```python

clf = tree.DecisionTreeClassifier(min_samples_leaf=10)


clf.fit(X_train, Y_train)
clf.score(X_test,Y_test)

Y_predicted = clf.predict(X_test)

tots = 0
for i in range(len(Y_test)):
    if Y_predicted[i] != Y_test[i]:
        tots +=1
print("The percentage of predictions that are not the same is {:.2f}%".format(tots/len(Y_predicted)*100))
```

```python
ids = pd.read_csv('test.csv')
ids = ids['surveyid']
ids = pd.DataFrame(ids)
Y_predicted = pd.DataFrame(Y_predicted)
Y_predicted.columns = ['depressed']

sub = pd.merge(ids,Y_predicted,left_index =True,right_index=True)

sub['depressed'] = sub['depressed'].astype(int)
sub.to_csv('andres_parrado_submission1.csv', encoding='utf-8')
```
